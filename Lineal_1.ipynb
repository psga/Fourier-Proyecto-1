```python
def vector_plot(tvects,is_vect=True,orig=[0,0,0]):
    """Plot vectors using plotly"""

    if is_vect:
        if not hasattr(orig[0],"__iter__"):
            coords = [[orig,np.sum([orig,v],axis=0)] for v in tvects]
        else:
            coords = [[o,np.sum([o,v],axis=0)] for o,v in zip(orig,tvects)]
    else:
        coords = tvects

    data = []
    for i,c in enumerate(coords):
        X1, Y1, Z1 = zip(c[0])
        X2, Y2, Z2 = zip(c[1])
        vector = go.Scatter3d(x = [X1[0],X2[0]],
                              y = [Y1[0],Y2[0]],
                              z = [Z1[0],Z2[0]],
                              marker = dict(size = [0,5],
                                            color = ['blue'],
                                            line=dict(width=5,
                                                      color='DarkSlateGrey')),
                              name = 'Vector'+str(i+1))
        data.append(vector)

    layout = go.Layout(
             margin = dict(l = 4,
                           r = 4,
                           b = 4,
                           t = 4)
                  )
    fig = go.Figure(data=data,layout=layout)
    fig.show()
a=np.array([0,2,0])
b=np.array([3,2,1])
c=np.array([1,-2,1])
d=np.array([-1,-1,-2])
```

```python
vector_plot([a,b,c,d])
```

```python
def v_a(a):
  if a==90:
    return "right"
  elif a>90 and a<180:
    return "obtuse"
  elif a<90 and a>0:
    return "acute"

z=[
    [np.array([2,1]),np.array([1,-3])],
    [np.array([2,-1,1]),np.array([1,-2,1])],
    [np.array([5,4,-3]),np.array([1,-2,-1])],
    [np.array([1,2,3,4]),np.array([-3,1,2,-2])]
  ]
for i in z:
  
  angle= (180/np.pi)*np.arccos(
          (np.dot(i[0],i[1]))/
          ( np.sqrt(
              np.dot(i[0],i[0])*
              np.dot(i[1],i[1]) 
              )
          )
      )

  print("Vectors {}\t and {}\t :the angle is {} // {}". format(
    i[0],
    i[1],
    angle,
    v_a(angle)     
  ))
```
